```python
import cv2

# Load the image
image = cv2.imread("images/lenna.png")

# Resize the image to a fixed width and height
resized_image = cv2.resize(image, (100, 200))

# Show the resized image
cv2.imshow("Resized Image", resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# Load the image
image = cv2.imread("images/lenna.png")

#Resize the image with aspect ratio preserved
resized_image = cv2.resize(image, None, fx=1.75, fy=1.75)

# Show the resized image
cv2.imshow("Resized Image", resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

###### Using skimage

```python
from skimage.transform import resize
import matplotlib.pyplot as plt

# Load the image
image = plt.imread("images/lenna.png")

# Resize with bilinear interpolation
resized_image = resize(image, (1000, 1200), anti_aliasing=True)

# Resize with bicubic interpolation
# resized_image = resize(image, (1300, 1300), anti_aliasing=True, order=3)

# Show the resized image
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(resized_image)
```
