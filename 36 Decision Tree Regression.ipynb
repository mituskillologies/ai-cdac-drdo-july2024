##### Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Load the data

```python
df = pd.read_csv('datasets/petrol_consumption.csv')
```

```python
df.shape
```

```python
df
```

##### Seperate the input and output data

```python
x = df.drop('Petrol_Consumption', axis = 1)
y = df['Petrol_Consumption']
```

```python
x.corrwith(y)
```

```python
sns.kdeplot(y)
```

```python
x.describe()
```

##### Cross Validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
x_train.shape, x_test.shape
```

##### Build the model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor(random_state=0)
```

```python
regressor.fit(x_train, y_train)
```

##### Performance Evaluation

```python
y_pred = regressor.predict(x_test)
```

```python
result = pd.DataFrame({'Actual':y_test, 'Predicted': y_pred})
result
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

##### Visualize

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(16,12))
plot_tree(regressor, fontsize=6, feature_names=x.columns,
         filled=True, precision=2, rounded=True);
```

```python
plt.title('Feature Importances')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.bar(x.columns, regressor.feature_importances_, color='purple')
```

##### Prediction on new / unseen data

```python
new = pd.DataFrame(data = [[7.3, 4449, 46039, 0.626],
                           [6.4, 5215, 23002, 0.672]],
                   columns=x.columns)
```
